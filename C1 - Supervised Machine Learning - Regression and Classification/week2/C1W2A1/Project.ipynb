```python
import numpy as np
import matplotlib.pyplot as plt
from utils import *
import copy
import math
%matplotlib inline
```

```python
x, y = load_data()
m = x.shape[0]
```

```python
plt.scatter(x, y, marker='x', color='g')
```

```python
def f(x, arg_l):
    return (x * arg_l[0] +  arg_l[1])
```

```python
def cost(x, y, arg_l, m):
    return np.sum((1 / (2*m)) * ((f(x, arg_l) - y) ** 2))
```

```python
def compute_gradient(x, y, arg_l, a, m):
    dj_dw = np.sum((1/m) * (f(x, arg_l) - y) * x)
    dj_db = np.sum((1/m) * (f(x, arg_l) - y))
    return [arg_l[0] - a*dj_dw, arg_l[1] - a*dj_db]
```

```python
c = 1
c_his = np.zeros((10000, 2))
count = 0
arg_l = [0, 0]
a = 5e-3
while count < 9999:
    arg_l = compute_gradient(x, y, arg_l, a, m)
    c = cost(x, y, arg_l, m)
    c_his[count] = np.array([count, c])
    count += 1
print(arg_l[0], arg_l[1])
plt.scatter(c_his[:,0], c_his[:,1])
```
